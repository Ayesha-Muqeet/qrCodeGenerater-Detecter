```python




!apt-get install -y libzbar0

!pip install qrcode[pil] opencv-python-headless pyzbar pillow
# ===============================
# QR Code Scanner & Generator
# By Ayesha
# ===============================

import cv2
import qrcode
from pyzbar import pyzbar
from PIL import Image
import datetime
import os
import webbrowser

from google.colab import files


# ========== History File ==========
HISTORY_FILE = "scan_history.txt"


# ========== Save History ==========
def save_history(data):
    time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    with open(HISTORY_FILE, "a", encoding="utf-8") as f:
        f.write(f"{time} | {data}\n")


# ========== Security Check ==========
def is_safe_link(data):
    if data.startswith("http"):
        bad_words = ["virus", "hack", "malware", "spam"]
        for word in bad_words:
            if word in data.lower():
                return False
        return True
    return True


# ========== Generate QR ==========
# ========== Generate QR ==========
def generate_qr():
    from IPython.display import display  # To show QR in Colab

    text = input("Enter text/link for QR: ")

    if text.strip() == "":
        print("‚ùå Empty input not allowed!")
        return

    # Create QR code
    qr = qrcode.QRCode(
        version=1,  # size of the QR Code
        box_size=10,
        border=5
    )
    qr.add_data(text)
    qr.make(fit=True)

    img = qr.make_image(fill_color="black", back_color="white")

    # Show QR code in Colab
    display(img)

    # Save QR code
    filename = "generated_qr.png"
    img.save(filename)

    # Download file
    print("‚úÖ QR Generated:", filename)
    files.download(filename)



# ========== Scan from Image ==========
def scan_image():
    print("üì§ Upload QR Image...")
    uploaded = files.upload()

    if not uploaded:
        print("‚ùå No file uploaded!")
        return

    for file in uploaded.keys():

        img = cv2.imread(file)

        decoded = pyzbar.decode(img)

        if not decoded:
            print("‚ùå No QR Code Found!")
            return

        for obj in decoded:

            data = obj.data.decode("utf-8")

            print("\n‚úÖ Scanned Data:", data)

            save_history(data)

            # Auto open link
            if data.startswith("http"):

                if is_safe_link(data):
                    print("üåê Opening Link...")
                    webbrowser.open(data)
                else:
                    print("‚ö†Ô∏è Warning: Unsafe Link!")


# ========== View History ==========
def view_history():

    if not os.path.exists(HISTORY_FILE):
        print("üìÅ No history found!")
        return

    print("\n===== SCAN HISTORY =====\n")

    with open(HISTORY_FILE, "r", encoding="utf-8") as f:
        print(f.read())


# ========== Main Menu ==========
def main():

    while True:

        print("\n==============================")
        print("   QR Code Project - Ayesha")
        print("==============================")
        print("1. Generate QR Code")
        print("2. Scan QR from Image")
        print("3. View Scan History")
        print("4. Exit")
        print("==============================")

        choice = input("Enter choice (1-4): ")

        if choice == "1":
            generate_qr()

        elif choice == "2":
            scan_image()

        elif choice == "3":
            view_history()

        elif choice == "4":
            print("üëã Goodbye!")
            break

        else:
            print("‚ùå Invalid Choice!")


# ========== Run ==========
main()
```
